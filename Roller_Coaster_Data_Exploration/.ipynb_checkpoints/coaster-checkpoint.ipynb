##### Data Preparation/Cleaning

```python
# Now, a data subset is created using the columns function so that only the necessary cloumns are reflected in the df, addding []

df.head ()
```

```python
#df.columns            #Here, I copied all columns from df.columns and unselected columns I don not wish to have for only 14 columns. Other way is to use drop ()
df = df[['coaster_name',
        #'Length', 'Speed', 
        'Location', 'Status', 
        #'Opening date',
         #'Type', 
        'Manufacturer', 
        #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
        #'Opened', 
        #'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
        'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
        #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
        #'height_value', 'height_unit', 
        'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()                                  #copy is needed so pandas knowss its a new df rather than a reference to previous one 
```

```python
df
```

```python
df.dtypes
```

```python
# Changing opening date to datetime

df ['opening_date_clean'] = pd.to_datetime (df['opening_date_clean'])
```

```python
df.dtypes
```

```python
# Lets rename the columns for ease and better structure 

df.rename (columns = { 'coaster_name' : 'Coaster_Name',
                         'year_introduced' : 'Year_Introduced',
                         'latitude' : 'Latitude',
                         'longitude' : 'Longitude',
                         'opening_date_clean' : 'Opening_Date',
                         'speed_mph' : 'Speed_mph',
                         'height_ft' : 'Height_ft',
                         'Inversions_clean' : 'Inversions',
                         'Gforce_clean' : 'Gforce'}, inplace = True)       # inplace to apply change in df
```

```python
# Identify null values

df.isnull ().sum()  
```

```python
# Identify duplicate values

#df.duplicated ()              #Returns true and false for the indexes, use of loc thus preferred to see them in df

df.loc [df.duplicated ()]
```

```python
#Lets check the duplicacy in the subset of columns

df.loc [df.duplicated (subset = ['Coaster_Name']) ]   #loc to specify location rather than labels or position
```

```python
# Lets see the data in coaster_name for why they are duplicated

df.query (' Coaster_Name == "Crystal Beach Cyclone" ')
```

```python
# Now, lets drop the second vaues that are duplicated coz of year_introduced and only keep as per name,location and opening date while dropping
# the second duplicated values in new df using inverse ~ to select 1st value and reset new index too

df.loc [~df.duplicated (subset = ['Coaster_Name', 'Location', 'Opening_Date']) ]
```

```python
# Here, the index isnt sequential anymore, so add the reset and assign new df so others are no longer in new df
# needs drop =True, so that the default new index column is removed

df = df.loc [~df.duplicated (subset = ['Coaster_Name', 'Location', 'Opening_Date']) ] \
        .reset_index (drop = True)
```

```python
# New shape

df.shape

#Hence, we dropped 97 duplicate columns in this df
```

##### Feature Understanding (Univariate Analysis)

```python
df['Year_Introduced'].value_counts()   # Count the values per year
```

```python
# Head to retrieve certain number of years only, 10 in this case

ax = df['Year_Introduced'].value_counts () \
    .head (10) \
    .plot (kind = 'bar', title = 'Top Years for Coaster Introduced', color = '#a64dff')
ax.set_xlabel ('Year Introduced')
ax.set_ylabel ('Number of Coasters')
plt.show()
```

```python
ax = df['Speed_mph'].plot(kind = 'hist', bins = 25, title = 'Roller Coaster Speed(mph)', color = '#a64dff')
ax.set_xlabel ('Speed (mph)')
plt.show()
```

```python
ax = df['Height_ft'].plot(kind = 'kde', title = 'Roller Coaster Height(ft)', color = '#a64dff')
ax.set_xlabel ('Height (ft)')
plt.show()
```

```python
type_count = df['Type_Main'].value_counts() 
type_count
```

##### Feature Relationships

```python
ax = df.plot (kind = 'scatter', x =  'Speed_mph', y = 'Height_ft', title = 'Coaster Speed vs Height')
ax.set_xlabel ('Speed (mph)')
ax.set_ylabel ('Height (ft)')
plt.show ()
```

```python
# Lets do the same with added features available using seaborn

ax = sns.scatterplot (x = 'Speed_mph', y = 'Height_ft', hue = 'Type_Main', data = df)
ax.set_title ('Coaster Speed vs. Height')
plt.show ()
```

```python
# Lets also use seaborn to compare multiple (>2) features. This creates a MAtrics like confusion_matrix

ax = sns.pairplot ( vars = ['Year_Introduced', 'Height_ft', 'Speed_mph', 'Inversions', 'Gforce'], data = df, hue = 'Type_Main')
plt.show ()
```

```python
# Lets also see the correlation between them. We drop the null values. This is applied on numeric values

corr = df [['Year_Introduced', 'Height_ft', 'Speed_mph', 'Inversions', 'Gforce']].dropna().corr ()
corr
```

```python
# Visually present the correlation in heatmap

corr_map = sns.heatmap (corr, annot = True)
```
