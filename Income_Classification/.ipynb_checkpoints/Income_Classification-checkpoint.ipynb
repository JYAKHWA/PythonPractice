##### This project as an experiment to classify the income into two target values as (>50K) and (<=50K).
##### The datset is obtained from UCI Machine Learning Repository, extracted from https://archive.ics.uci.edu/dataset/2/adult
##### The dataset is obtained in a text file, which is convertedd into a csv and added columns as described in the UCI Repository, having 14 Attributes, 48K+ instances containing both integer and categorical data, classified as a multivariate dataset.


#### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

import tensorflow as tf
from tensorflow.keras.models import Sequential

import warnings
warnings.filterwarnings ('ignore')
```

#### Import and load converted csv dataset

```python
#data = np.load ('adult.csv', allow_pickle = True)

#pd.set_option ('display.max_rows', None)

data = pd.read_csv ('adult.csv')
data.head (8)
```

##### We consider education and fnlwgt having no meaning and thus we drop it.

```python
# Selecting the columns to include in the data and coping them, instead of dropping the columns
data = data [['age', 'workclass',
              'education-num',
           'marital-status', 'occupation', 'relationship', 'race', 'sex',
           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
           'class']].copy()
```

```python
data.head (2)
```

##### Changing the target name class to income

```python
data = data.rename (columns = {'class' : 'income'} )
data.head (2)
```

##### Here, all columns have values mentioned '?'. Lets replace it will NaN so that NaN can then be filled up with ffill (Forward Fill)

```python
data ['workclass'].unique()
```

```python
data.replace ('?', np.NaN, inplace = True)
data.head ()
```

##### Lets strip the spaces as it isn't working

```python
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)

data.replace ('?', np.NaN, inplace = True)
```

```python
data ['occupation'].unique()
```

##### ? has now been replaced by NaN. Lets fill the NaN with ffill

```python
data.fillna (method = 'ffill', inplace = True)
```

```python
data ['occupation'].unique()
```

#### Feature Enginnering
##### Education column has 7 categories. coberting them to 3

```python
data ['marital-status'] = data['marital-status'].replace ( ['Married-civ-spouse', 'Married-AF-spouse'], 'Married' )
data ['marital-status'] = data['marital-status'].replace ( ['Never-married'], 'Not-Married' )
data ['marital-status'] = data['marital-status'].replace ( ['Married-spouse-absent', 'Divorced', 'Separated', 'Widowed'], 'Others' )
```

#### Label Encoding

```python
le = LabelEncoder ()
```

```python
# Copy data to new df and encode the cols
df = data.copy ()
df = df.apply (le.fit_transform)
df.head()
```

```python
corr = df.corr()
htmap = sns.heatmap (corr, cmap = 'inferno')
htmap.set_title ('Correlation Heatmap')
htmap;
```

#### Data Modelling

```python
x = df.drop (['income'], axis = 1)
y = df['income']
```

##### Normalization


```python
scaler = StandardScaler().fit (df.drop('income', axis = 1))
scaler
```

```python
X = scaler.transform (df.drop ('income', axis = 1) )
y = df['income']
```

#### Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split (X, y, test_size = 0.2, random_state = 50)
```

#### Logistic Regresssion (Binary Classification)

```python
lr = LogisticRegression ()

model = lr.fit (x_train, y_train)
prediction = model.predict (x_test)
print ('Accuracy on training data: {:,.2f}'.format(lr.score(x_train, y_train)))
print ('Accuracy on testing data: {:,.2f}'.format(lr.score(x_test, y_test)))
```

#### Random Forest

```python
ranfor = RandomForestClassifier ()
model1 = ranfor.fit (x_train, y_train)
prediction1 = model1.predict (x_test)
print ('Accuracy on training data: {:,.2f}'.format(ranfor.score(x_train, y_train)))
print ('Accuracy on testing data: {:,.2f}'.format(ranfor.score(x_test, y_test)))
```

#### Naive Bayes

```python
gb = GaussianNB ()
model2 = gb.fit (x_train, y_train)
prediction2 = model2.predict (x_test)
print ('Accuracy on training data: {:,.2f}'.format(gb.score(x_train, y_train)))
print ('Accuracy on testing data: {:,.2f}'.format(gb.score(x_test, y_test)))
```

#### Artificial Neural Network (ANN)

```python
model_nn = Sequential ( [
    tf.keras.layers.Dense (32, activation = 'relu', input_shape = x_train.shape[1:]),
    tf.keras.layers.Dense (32, activation = 'relu'),
    tf.keras.layers.Dense (1, activation = 'sigmoid')
] )
```

```python
model_nn.compile (optimizer = 'sgd',
                  loss = 'binary_crossentropy',
                  metrics = 'accuracy')
```

```python
hist = model_nn.fit (x_train, y_train, batch_size = 32, epochs = 20, validation_data = (x_test, y_test))
```

```python
model_nn.summary()
```

```python
# Predictions and calculation of accuracy
predict = model_nn.predict(x_test)
result = model_nn.evaluate(x_test, y_test)

print(f'Test set loss: {result[0]:0.3f}, test set accuracy: {result[1]*100:0.2f}%')
```

##### Final Accuracy Result:
1. Logistic Regression: 85%
2. Random Forest: 86%
3. Naive Bayes: 82%
4. Artificial Neural Network (ANN): 85.67%

```python
# Confusion_Matrix from RF as it has the highest accuracy
con_matrix = print (confusion_matrix(np.ravel(y_test), np.ravel(prediction1)))
con_matrix
```
